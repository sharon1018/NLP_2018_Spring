```python
import re
from collections import Counter
from nltk import ngrams

def words(text): return re.findall(r'\w+', text.lower())

WORDS = Counter((open("C:/Users/asus/Downloads/nlp/big.txt", encoding='utf-8').read()))
def P(word, N=sum(WORDS.values())):  #N: sum of all words
    "Probability of `word`."
    return WORDS[word] / N 
```

```python
def next_states(state):
    letters = 'abcdefghijklmnopqrstuvwxyz'
    L, R, edit, prob = state
    R0, R1 = R[0], R[1:]
    all_word=[]
    if edit<=2: 
        all_word.append((    L + R0,     R1, edit,   P(L+R0+R1) ))      #noedit
        all_word.append((    L ,         R1 ,edit+1, P(L+R1)))          #delete
        for c in letters:
            all_word.append((L + c ,     R1, edit+1, P(L+c+R1)))        #replace
            all_word.append((L + R0 + c, R1, edit+1, P(L+R0+c+R1)))     #insert      
        return all_word

word='appearant'                     
# print(next_states(('', word ,0, P(word))))
```
